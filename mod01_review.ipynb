```python
import pandas as pd
import numpy as np

# Load data
Credit = pd.read_csv('data/Credit.csv')

# Feature names
print("\nFeatures:")
print(Credit.columns)

# Data types
print("\nData Types:")
print(Credit.dtypes)

# Age statistics
print("\nAge Statistics:")
print("Mean:", Credit['Age'].mean())
print("Median:", Credit['Age'].median())
print("Mode:", Credit['Age'].mode().values)

# Create subset
Subset = Credit[['Age', 'Region', 'Income']]

print("\nSubset Preview:")
print(Subset.head())

# Partition by age
Younger = Subset[Subset['Age'] < 50]
Older = Subset[Subset['Age'] >= 50]

# Income comparison
print("\nIncome Comparison (Age < 50 vs Age 50+):")
print("Younger Mean Income:", Younger['Income'].mean())
print("Older Mean Income:", Older['Income'].mean())
print("Younger Median Income:", Younger['Income'].median())
print("Older Median Income:", Older['Income'].median())

# Region values
print("\nUnique Regions:")
print(Credit['Region'].unique())

# Region frequency table
print("\nRegion Frequency Table:")
print(Credit['Region'].value_counts())

# Student vs Non-student income
print("\nAverage Income by Student Status:")
print(Credit.groupby('Student')['Income'].mean())

# Add utilization feature
Credit['Utilization'] = Credit['Balance'] / Credit['Limit']

print("\nUtilization Preview:")
print(Credit[['Balance', 'Limit', 'Utilization']].head())

# Maximum utilization
print("\nMaximum Utilization Value:")
print(Credit['Utilization'].max())
```
